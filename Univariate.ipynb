```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("train.csv")

# Optional: label mappings for better readability
label_mappings = {
    "Gender": {0: "Male", 1: "Female"},
    "HiringDecision": {0: "Not Hired", 1: "Hired"},
    "EducationLevel": {1: "Bachelor's(Type 1)", 2: "Bachelor's(Type 2)", 3: "Master's", 4: "PhD"},
    "RecruitmentStrategy": {1: "Aggressive", 2: "Moderate", 3: "Conservative"}
}

# Define categorical columns
categorical_columns = ["Gender", "EducationLevel", "RecruitmentStrategy", "HiringDecision"]

# Set seaborn style
sns.set(style="whitegrid")

# Define custom color palette
color_palette = ["#ADD8E6", "#4169E1", "#4682B4", "#20B2AA"]

# Generate separate charts for each variable
for col in categorical_columns:
    plt.figure(figsize=(6, 6))

    # Map labels if available
    if col in label_mappings:
        data = df[col].map(label_mappings[col])
    else:
        data = df[col]

    # Pie Chart
    data_counts = data.value_counts()
    data_counts.plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        colors=color_palette[:len(data_counts)],
        ylabel=''
    )
    plt.title(f'Pie Chart of {col}')
    plt.tight_layout()
    plt.show()

    # Bar Chart
    plt.figure(figsize=(8, 4))
    sns.countplot(x=data, palette=color_palette[:len(data_counts)], hue=data, legend=False)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```
