```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv(r"C:\Users\mpkhd\Desktop\Uni\4th year\Sem1\ST4052\Project1\EDA\train.csv")
df = df.astype({'Gender': 'str','PreviousCompanies':'str','EducationLevel':'str',
                'RecruitmentStrategy':'str', 'HiringDecision':'str'})
```

```python
df.dtypes
```

```python
import scipy.stats as stats
from scipy.stats import skew
from scipy.stats import kurtosis

stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title('Q–Q Plot')
plt.show()

print("Skewness:", skew(df['Age']))
print("Excess Kurtosis:", kurtosis(df['Age']))  

res = stats.shapiro(df['Age'])
print("Shapiro P value: ",res.pvalue)
```
